```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df
```

```python
df.drop(columns=['Director', 'Star1', 'Star2', 'Star3', 'Star4'], inplace=True)
```

```python
df
```

```python
df.drop(columns=['Poster_Link'], inplace=True)
df
```

```python
df.rename(columns={'Series_Title': 'Title'}, inplace=True)
top_10 = df[['Title', 'IMDB_Rating']].sort_values(by='IMDB_Rating', ascending=False).head(10)
top_10
```

```python
plt.figure(figsize=(15,6))
plt.barh(top_10['Title'], top_10['IMDB_Rating'])
plt.xlabel('IMDB_Rating')
plt.title('Top 10 Movies by IMDB Rating')
plt.show()
```
